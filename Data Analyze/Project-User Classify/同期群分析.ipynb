```python
import pandas as pd
import os
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('同期群分析实战案例数据.xlsx')
df.info()
```

```python
df.head()
```

```python
#calculate by month, so extract the month
df['Month'] = pd.to_datetime(df['付款时间'].apply(lambda x:x[:7]))
df.head()
```

```python
df['First_Month'] = df.groupby('用户ID')['Month'].transform('min')
df['Gap'] = df['Month'] - df['First_Month']
df['Gap'] = df['Gap'].apply(lambda x:x.days)
df['Gap'] = df['Gap'].apply(lambda x:round(x/30))
df_group = df.groupby(['First_Month', 'Gap'], as_index=False)['用户ID'].apply(pd.Series.nunique)
df_group
```

```python
df_pivot = df_group.pivot(index='First_Month', columns='Gap', values='用户ID')
df_pivot
```

```python
save_rate = df_pivot.divide(df_pivot.iloc[:,0], axis=0).round(3)*100
save_rate
```

```python
save_rate.iloc[:,0] = df_pivot.iloc[:,0]
save_rate
```

```python
save_rate.index = [str(i)[:7] for i in save_rate.index.to_list()]
save_rate.columns=['当月新增','+1月','+2月','+3月','+4月','+5月','+6月','+7月','+8月','+9月','+10月','+11月']
save_rate
```

```python
#money cohort
df_group_money = df.groupby(['First_Month', 'Gap'], as_index=False)['实付金额'].sum()
df_pivot_money = df_group_money .pivot(index='First_Month', columns='Gap', values='实付金额')
df_pivot_money.index = [str(i)[:7] for i in df_pivot_money.index.to_list()]
df_pivot_money.columns=['当月新增','+1月','+2月','+3月','+4月','+5月','+6月','+7月','+8月','+9月','+10月','+11月']
df_pivot_money
```

```python
df_pivot.index = [str(i)[:7] for i in df_pivot.index.to_list()]
df_pivot.columns=['当月新增','+1月','+2月','+3月','+4月','+5月','+6月','+7月','+8月','+9月','+10月','+11月']
avg_money = round(df_pivot_money/df_pivot)
avg_money['当月新增'] = save_rate['当月新增']
avg_money
```
