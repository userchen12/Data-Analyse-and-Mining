```python
import os
os.chdir(path)
df = pd.DataFrame()
for file in os.listdir():
    data = pd.read_excel(file)
    data['Date'] = file[:4]
    df = pd.concat([df, data])

df = df.loc[df['Date'].str.find('2023') != -1,:]
print('data rows: {}'.format(len(df)))
```

```python
df.head()
```

```python
df = df.groupby('品牌')[['品牌搜索人数','点击人数','支付人数']].sum().reset_index().sort_values('品牌搜索人数', ascending=False)
df.head()
```

```python
#calculate the transform rate
df['SearchClickRate'] = df['点击人数']/df['品牌搜索人数']
df['ClickPayRate'] = df['支付人数']/df['点击人数']
df.head()
```

```python
#only filter top15
data = df.copy()
data = data.iloc[:15,:]

plt.figure(figsize=(12,7))

x = data['SearchClickRate'].to_list()
y = data['ClickPayRate'].to_list()
z = data['品牌搜索人数']

text = data['品牌'].to_list()

#draw
plt.scatter(x, y, s=z/1000, c=x, cmap='Reds', alpha=0.7, edgecolors='grey', linewidths=1)

#add brand name
for i, txt in enumerate(text):
    plt.text(x=x[i], y=y[i], s=txt, size=11, horizontalalignment='center', verticalalignment='center')
    
#add line
plt.axhline(y=0.12, color='black', linestyle='--', alpha=0.5)
plt.axvline(x=0.465, color='black', linestyle='--', alpha=0.5)

plt.xlabel('SearchClickRate')
plt.ylabel('ClickPayClickRate')
plt.title('Top15 Advertising Effect')
    
```
