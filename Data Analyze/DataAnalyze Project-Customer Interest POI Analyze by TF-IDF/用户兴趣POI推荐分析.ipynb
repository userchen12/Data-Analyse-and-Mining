2.1 用户POI基础权重模型

用户的POI权重模型采取TF-IDF（terms frequency - inverse document frequency），其中TF标示了此POI在用户所有POI中的重要程度，它是此POI的频数在当前用户所有POI频数所占的比例；IDF标示了POI在所有用户产生的POI的热门程度，这表达了一种常识意义上的直觉，如果一个POI出现在所有用户中,那么这个POI在区分出用户的POI方面几乎不起任何作用，而相反地，如果一个POI是相对冷门一些的兴趣点，那么它更能体现用户真正的兴趣，例如热门“影视”和相对冷门（与前者相比）一些的“美国功夫大片”，那么后者更能体现用户具体的兴趣。可以看出，TF是对每个用户而异的，而IDF是POI全局的性质，不会因为用户而发生变化

用户的权重计算采取TF*IDF权重模型：

Wi ( u, p) = TFi* IDF = TFi* log(N /ni + 0. 01)

其中TFi 为POI pi 在当前用户所有标注中出现的次数占比，称为项频，N为全部用户集合U中的用户总数，ni 为U中使用过POI pi 的用户数。这个公式的核心思想是一个POI被一个用户使用的频率越高，而被其他用户使用的频率越低，则该POI对用户兴趣的区分能力越强，故其权重也越大。事实上，如果一个用户经常使用某个词语作为POI，说明他对这个方面的兴趣是相对更加强烈的，持续时间也更长，同时，这个POI越冷门则越能体现这个用户兴趣的特殊性，因此对这种POI的权重更需要加强。

```python
import pandas as pd
df_movies = pd.read_csv(r'D:\Learning\Data Analyse\Project\数据集\movie dataset\movies.csv')
df_tags = pd.read_csv(r'D:\Learning\Data Analyse\Project\数据集\movie dataset\tags.csv')
```

```python
df_user_tags = pd.merge(df_tags, df_movies, on='movieId', how='left')[['userId','movieId','tag']]
df_user_genres = pd.merge(df_tags, df_movies, on='movieId', how='left')[['userId','movieId','genres']].drop_duplicates()
df = (
    df_user_genres.assign(genres=df_user_genres['genres'].str.split('\\|'))
    .explode('genres')
    .reset_index(drop=True)
)
df.head()
tag_col = 'genres'
```

```python
#计算TF-IDF，
import numpy as np
df['tag_count'] = df.groupby(['userId'])[tag_col].transform('count')
df['user_tag_count'] = df.groupby(['userId',tag_col])[tag_col].transform('count')
df['tag_TF'] = df['user_tag_count']/df['tag_count']

df['tag_user_count'] = df.groupby([tag_col])['userId'].transform('nunique')
df['users_count'] = df['userId'].nunique()
df['tag_IDF'] = np.log(df['users_count']/df['tag_user_count'] + 0.01)
df['TF-IDF'] = df['tag_TF']*df['tag_IDF']
df.head()
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#查看TIDF值分布
sns.scatterplot(x='tag_user_count', y='tag_IDF', data=df)
sns.lineplot(x='tag_user_count', y='tag_IDF', data=df)
plt.show()
```

```python
#挑选例子看TF值分布
user_4 = df[df['userId'] == 4]
data=user_4[[tag_col, 'tag_TF']].drop_duplicates().sort_values(by='tag_TF', ascending=False)
plt.bar(data[tag_col], data['tag_TF'],color='grey')
plt.xticks(rotation=90)
plt.show()
```

```python
#查看TF-IDF值分布
data=user_4[[tag_col, 'TF-IDF']].drop_duplicates().sort_values(by='TF-IDF', ascending=False)
plt.bar(data[tag_col], data['TF-IDF'],color='grey')
plt.xticks(rotation=90)
plt.show()
```

```python
#加入时间衰减因子
import math
df_user_tags = pd.merge(df_tags, df_movies, on='movieId', how='left')

df_user_tags =(
    df_user_tags.assign(genres=df_user_tags['genres'].str.split('\\|'))
    .explode('genres')
    .reset_index(drop=True)
)
# df_user_tags['timestamp'] = (df_user_tags['timestamp']/86400).astype('int')
df_user_tags['tag_recent'] = df_user_tags .groupby(['userId', tag_col])['timestamp'].transform('max')
df_user_tags['tag_first'] = df_user_tags .groupby(['userId', tag_col])['timestamp'].transform('min')
df_user_tags['user_recent'] = df_user_tags .groupby(['userId'])['timestamp'].transform('max')
df_user_tags['user_first'] = df_user_tags .groupby(['userId'])['timestamp'].transform('min')-1

#时间衰减是非线性的，类似sigmod函数
df_user_tags['decay'] = (df_user_tags['user_recent']-df_user_tags['tag_recent'])/(df_user_tags['user_recent']-df_user_tags['user_first'])
def sigmod_transfer(x):
    return 2/(1+math.exp(-x*100))
df_user_tags['decay'] = df_user_tags['decay'].apply(lambda x:sigmod_transfer(x))

df_re = pd.merge(df, df_user_tags, on=['userId', 'movieId','genres'], how='left')
df_re = df_re[['userId', 'movieId', 'tag','genres','tag_TF','tag_IDF', 'TF-IDF','decay']]
df_re['TF-IDF_re'] = df_re['TF-IDF']*df_re['decay']
df_re.head()
```
