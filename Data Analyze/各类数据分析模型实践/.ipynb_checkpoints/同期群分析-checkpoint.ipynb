```python
import numpy as np 
import pandas as pd 
import datetime as dt 
 
import matplotlib.pyplot as plt
import seaborn as sns
 
#设置汉字格式：Trebuchet MS, Tahoma, Verdana, Arial, Helvetica,SimHei 中文的幼圆、隶书等等

# 解决中文无法显示问题
plt.rcParams['font.sans-serif']=['Songti SC']   
# 解决保存图像中负号'-'显示为方块问题
plt.rcParams['axes.unicode_minus'] = False 

path = r'../../../数据集\eCommerce Events History in Cosmetics Shop/2019-Dec.csv'
df = pd.read_csv(path)
```

```python
# data clean
# select price > 0 and drop duplicates
data = df.query('price > 0').drop_duplicates()

# date format
def get_date(timestamp):
    date = dt.datetime.strptime(timestamp, "%Y-%m-%d %H:%M:%S %Z").date()
    return date

data['event_date'] = data['event_time'].apply(lambda x:get_date(x))
    
```

```python
# calculate the fisrt login in date of erery user
data['first_login_date'] = data.groupby(['user_id'])['event_date'].transform('min')
data['date_gap'] = data['event_date']-data['first_login_date']
data['date_gap'] = data['date_gap'].apply(lambda x:x.days)
data.head(5)
```

```python
data_group = data.groupby(['first_login_date', 'date_gap'], as_index=False)['user_id'].apply(pd.Series.nunique)
#create pivot
data_pivot = data_group.pivot(index='first_login_date', columns='date_gap', values='user_id')
data_pivot
```

一般来说，留存有两种算法：
算法1中是把当前日直接记为Day1，还有一种计算方法将当前日当做Day0，然后再开始计算；
- 如果使用算法1，7日日留存=下周1的数据/本周2的数据(10)
- 如果使用算法2，7日日留存=下周2的数据/本周2的数据(10)

算法2在一定程度上能够巧妙避开星期级别的影响

```python
# use the first  column as baseline
size = data_pivot.iloc[:,0]
save_rate = data_pivot.divide(size, axis=0)
```

```python
save_rate = save_rate.round(3)*100
save_rate.head(5)
```

```python
plt.rcParams['axes.unicode_minus'] = False 

plt.figure(figsize=(15,8))
plt.title("Cohort Result")

sns.heatmap(data=save_rate,
            annot=True,
            fmt='.0%',
            vmin = 0.0,
            vmax = 0.2,
            cmap="BuPu_r"
           )
plt.show()
```
